```python
import spacy
import re

nlp = spacy.load("en_core_web_sm")

def replace_addresses(text, replacement="<ADDRESS>"):
    """Detects and replaces full addresses while avoiding false positives (e.g., technologies)."""

    # âœ… Step 1: More precise regex patterns to detect addresses
    address_patterns = [
        r"\d{1,5}\s[\w\s]+,\s[\w\s]+,\s[A-Z]{2}\s\d{5}(-\d{4})?",  # "123 Main St, City, ST 12345"
        r"\d{1,5}\s[\w\s]+,\s[A-Z]{2}\s\d{5}(-\d{4})?",  # "123 Main St, ST 12345"
        r"\d{1,5}\s[\w\s]+,\s[\w\s]+,\s[A-Z]{2}",  # "123 Main St, City, ST"
        r"\d{1,5}\s[\w\s]+,\s[A-Z]{2}",  # "123 Main St, ST"
        r"\d{1,5}\s[\w\s]+",  # Basic "123 Main St" but not too aggressive
        r"[\w\s]+,\s[A-Z]{2}\s\d{5}(-\d{4})?",  # "City, ST 12345"
        r"\b\d{5}(-\d{4})?\b(?![-.\s]\d{3})",  # **Standalone ZIP codes but avoid breaking phone numbers**
    ]

    # Apply regex-based address replacements
    for pattern in address_patterns:
        text = re.sub(pattern, replacement, text)

    # âœ… Step 2: Use spaCy NER but filter out false positives
    doc = nlp(text)
    locations = set()

    for ent in doc.ents:
        if ent.label_ in ["GPE", "LOC", "FAC"] and ent.text not in false_positives:
            locations.add(ent.text)

    # Replace detected locations while avoiding false positives
    if locations:
        pattern = re.compile(r'\b(' + '|'.join(map(re.escape, locations)) + r')\b')
        text = pattern.sub(replacement, text)

    # âœ… Step 3: Smarter merging of consecutive `<ADDRESS>` placeholders
    text = re.sub(r"\[Your City, State\]", replacement, text)  # Keep existing manual replacements
    text = re.sub(r"\bFlorida\b(?!\s[A-Z]{2})", replacement, text)  # Replace "Florida" only if not followed by a state code
    text = re.sub(r'(<ADDRESS>[\.,]?\s?)+<ADDRESS>', replacement, text)  # Merge only consecutive `<ADDRESS>` instances

    return text

# âœ… Example: Test with problematic cases
text = """
Maintain <ADDRESS> web pages for Network Tools utilizing JavaScript, Node.js, CSS,
<ADDRESS>, and SQL Server.

Maintain HTML web pages for Network Tools utilizing JavaScript, Node.js, CSS,
Perl, PHP, and SQL Server.

The company headquarters is located at 123 Main St, Springfield, IL 62704.

Database work involves MongoDB and PostgreSQL.

Frontend tech includes React, Bootstrap, and Angular.

Visit our office at 55 Montgomery Rd., Cambridge, MA 02138.

Front-End: Javascript ES6, React, React Hooks, HTML5, CSS3, JSX
Back-End: Node.js, Express.js, Knex.js, PostgreSQL
Misc.: Mocha, Chai, Git, Github, Vercel, NPM, Adobe Photoshop / Creative Suite

"""

# Run function
modified_text = replace_addresses(text)

# Print output
print(modified_text)
```

```python
import spacy
import re

nlp = spacy.load("en_core_web_sm")

def replace_dates(text, replacement="<DATE>"):
    """
    Detects and replaces various date formats, including MM/YY and M/YY, using regex and spaCy NER.
    """

    # âœ… Step 1: Regex-based date replacement

    # 1ï¸âƒ£ Full & Short Month Names + Year (e.g., "January 2020", "Jan 2020")
    text = re.sub(r'\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sept?|Oct|Nov|Dec)\.? \d{4}\b', replacement, text)
    text = re.sub(r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December) \d{4}\b', replacement, text)

    # 2ï¸âƒ£ Numeric Date Formats (MM/DD/YYYY, DD-MM-YYYY, YYYY-MM-DD)
    text = re.sub(r'\b\d{1,2}[-/.]\d{1,2}[-/.]\d{2,4}\b', replacement, text)  # Captures "12/31/2022", "31-12-2022"
    text = re.sub(r'\b\d{4}[-/.]\d{1,2}[-/.]\d{1,2}\b', replacement, text)  # Captures "2022-12-31"

    # 3ï¸âƒ£ New: Capture MM/YY or M/YY (e.g., "3/04", "12/99")
    text = re.sub(r'\b\d{1,2}/\d{2}\b', replacement, text)

    # 4ï¸âƒ£ Month, Day, Year Format (e.g., "February 3, 2021")
    text = re.sub(r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December) \d{1,2}, \d{4}\b', replacement, text)

    # 5ï¸âƒ£ "From - To" Date Ranges (e.g., "March 2015 - April 2020")
    text = re.sub(r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December) \d{4}\s?[-â€“]\s?(?:January|February|March|April|May|June|July|August|September|October|November|December)?\s?\d{4}\b', replacement, text)

    # 6ï¸âƒ£ Standalone Four-Digit Years (excluding phone numbers)
    text = re.sub(r'(?<!\d{3}[-.\s])\b\d{4}\b(?![-.\s]\d{3})', replacement, text)
    text = re.sub(r'\(?\b\d{4}[-â€“]\d{4}\b\)?', replacement, text)  # Handles (2008-2012) and 2008-2012

    # 7ï¸âƒ£ Other common date references
    text = re.sub(r'Present', replacement, text)
    text = re.sub(r'Current', replacement, text)

    # 8ï¸âƒ£ Context-Based Dates (e.g., "Graduated in 2020", "Worked for 3 years")
    text = re.sub(r'\b(?:Graduated|Completed|Finished|Started|Since|From) in \d{4}\b', replacement, text)
    text = re.sub(r'\b(?:for|over|about) \d{1,2} years?(?: and \d{1,2} months?)?\b', replacement, text)

    # 9ï¸âƒ£ Decades (e.g., "The 1990s", "Early 2000s")
    text = re.sub(r'\b(?:the )?\d{4}s\b', replacement, text)

    # âœ… Step 2: Use spaCy's NER to detect and replace additional dates
    doc = nlp(text)
    dates = set(ent.text for ent in doc.ents if ent.label_ == "DATE")

    if dates:
        pattern = re.compile(r'\b(' + '|'.join(map(re.escape, dates)) + r')\b')
        text = pattern.sub(replacement, text)

    return text

# âœ… Example Usage
text = """
I was born on January 5, 1995. The project started on 12/10/2020 and ended on March 15, 2021.
She will visit on Monday, October 3rd. The contract expires in July 2024.
Graduated in 2018. Worked for 3 years. The 1990s were a great decade.
Started my job in 3/04 â€“ 17/22
"""

# Run function
modified_text = replace_dates(text)

# Print output
print(modified_text)
```

# ğŸ¯ Code for `<PHONE_NUMBERS>`

```python
def replace_phone_numbers(text):
    """
    Replaces all phone numbers and the exact string "[Your Phone Number]" with <PHONE_NUMBER>.

    Args:
        text (str): The input text.

    Returns:
        str: The text with phone numbers replaced by <PHONE_NUMBER>.
    """
    # âœ… Regex for real phone numbers
    phone_pattern = r'''
        (?:                                  # Non-capturing group for phone numbers
            (?:\+?\d{1,3}[-.\s]?)?           # Optional country code (+1, +44, etc.)
            (?:\(?\d{3}\)?[-.\s]?)           # Area code (with or without parentheses)
            \d{3}[-.\s]?\d{4}                # Main phone number (XXX-XXXX, XXX.XXXX, or XXX XXXX)
        )
    '''

    # âœ… First replace real phone numbers
    text = re.sub(phone_pattern, '<PHONE_NUMBER>', text, flags=re.VERBOSE)

    # âœ… Now separately replace the exact string "[Your Phone Number]"
    text = re.sub(r'\[Your Phone Number\]', '<PHONE_NUMBER>', text)

    return text

# # âœ… Test Cases
text = """
Call me at +1 (305) 555-1234 or (123) 456-7890.
Also, my placeholder is [Your Phone Number].
Another number: 987-654-3210.
My UK number is +44 20 7946 0958.
Reach me at 202.555.0173 or 555 867 5309.
Office: +91-98765-43210, Personal: (415)888-9999.
We can be contacted at 5551234 or 555-123-4567.
Emergency? Call 1-800-123-4567 or (800) 555-1212.
Invalid: 123-45-678 (should not be replaced).
"""

print(replace_phone_numbers(text))

```

# ğŸ¯ Code for `<LINKS>`


---
Detects and replaces URLs and links with <LINK>, while avoiding emails and false positives.

    Handles:
    - "https://example.com"
    - "http://example.org/page"
    - "www.example.com"
    - "example.com/profile"
    - "github.com/johndoe"

Avoids:
    - Emails like "john.doe@example.com"
    - Words containing ".com" that are not links (e.g., "E-commerce")



---



```python
def replace_links(text):
    """
    Detects and replaces valid links while avoiding emails and other non-URL text.
    """

    # âœ… Step 1: Remove "unavailable" placeholders
    text = text.replace("unavailable)", "").replace("unavailable", "")

    # âœ… Step 2: Replace any string containing "www."
    text = re.sub(r'\bwww\.[^\s,"]+', '<LINK>', text)
    text = re.sub(r'\[Your LinkedIn Profile URL (Optional)\]', '<LINK>', text)

    # âœ… Step 3: Improved regex for detecting valid links while avoiding emails
    link_pattern = r'''
        (?<!\S)  # Ensure it's not part of another word
        (
            https?://[^\s,"]+  |   # Matches "http://example.com", "https://example.com/path"
            [a-zA-Z0-9.-]+\.(com|net|org|edu|gov|io|ai)(/[^\s,"]*)?  # Matches "example.com", "example.com/path" (but avoids emails)
        )
    '''

    # âœ… Step 4: Replace detected links
    text = re.sub(link_pattern, '<LINK>', text, flags=re.VERBOSE)

    # âœ… Step 5: Replace placeholders like "[Your LinkedIn Profile URL]" and variations of "(link)"
    placeholder_patterns = [
        r'\[Your LinkedIn Profile URL\]',
        r'\[link unavailable\]',
        r'\[link\]',
        r'\(\s*[^)]*\blink\b[^)]*\s*\)'  # Matches variations like "(See the Link for details)"
    ]

    for pattern in placeholder_patterns:
        text = re.sub(pattern, '<LINK>', text, flags=re.IGNORECASE)

    return text

# âœ… Test Cases
text = """
(link unavailable) Check out my profile (link provided here)." (Link)
"Visit my portfolio at https://myportfolio.com."
"www.github.com/johndoe is my GitHub."
"(See the Link for details)"
"[Your LinkedIn Profile URL] is missing."
"Email me at john.doe@example.com (not a link)."
"(click the link below for info)"
"Check my website at example.com/about."
"""

print(replace_links(text))

```

```python
def replace_universities(text):
    """
    Detects and replaces university, college, and institute names with <INSTITUTION>.

    Handles:
    - "Graduated from Stanford University" â†’ "Graduated from <INSTITUTION>"
    - "Studied at the University of California" â†’ "Studied at <INSTITUTION>"
    - "Earned a degree from Harvard College" â†’ "Earned a degree from <INSTITUTION>"
    - "Attended UCLA" â†’ "Attended <INSTITUTION>"
    - "Massachusetts Institute of Technology" â†’ "<INSTITUTION>"
    - "Graduated from Oxford Academy" â†’ "<INSTITUTION>"

    Args:
        text (str): The input text containing university names.

    Returns:
        str: The text with universities replaced by <INSTITUTION>.
    """

    # âœ… Merged regex pattern for universities, colleges, institutes, polytechnics, academies
    university_pattern = r'\b(?:' + '|'.join([
        r'University of [A-Z][a-z]+(?:\s[A-Z][a-z]+)*',  # "University of Michigan"
        r'UNIVERSITY OF [A-Z]+(?:\s[A-Z]+)*',
        r'[A-Z][a-z]+ University',  # "Stanford University"
        r'[A-Z][a-z]+ Institute of Technology',  # "Massachusetts Institute of Technology"
        r'[A-Z][a-z]+ Institute',  # "California Institute"
        r'[A-Z][a-z]+ College',  # "Harvard College"
        r'[A-Z][a-z]+ Polytechnic',  # "XYZ Polytechnic"
        r'[A-Z][a-z]+ Academy',  # "Oxford Academy"
        r'[A-Z][a-z]+ School of [A-Z][a-z]+',  # "London School of Economics"
        r'UCLA|MIT|NYU|Caltech|Stanford|Harvard|Yale|Princeton|Oxford|Cambridge'  # Specific university abbreviations
    ]) + r')\b'

    # âœ… Replace detected university names
    text = re.sub(university_pattern, '<INSTITUTION>', text)

    return text

text = """
I graduated from Stanford University with a degree in Computer Science.
THE UNIVERSITY OF NORTH CAROLINA AT GREENSBORO
university of north carolina at greensboro
My sister studied at the university of California, Berkeley.
He earned his Ph.D. from the Massachusetts Institute of Technology.
She completed her law degree at Harvard College.
John attended UCLA for his undergraduate studies.
Cambridge University has a rich history of academic excellence.
The Oxford Academy is known for its rigorous curriculum.
I was admitted to Princeton, but later transferred to Yale.
She went to the London School of Economics to study finance.
NYU offers a great film program, while Caltech is known for engineering.
I did a short course at XYZ Polytechnic last summer.
The California Institute of Technology is ranked among the best.
"""
print(replace_universities(text))
```

#âœŒï¸ Function to Extract Text

```python
import os
import pdfplumber
import docx
from google.colab import drive
```

```python
def extract_text(filepath):
    """Extracts text from .pdf, .docx, and .txt files."""
    text = ""
    if filepath.endswith(".pdf"):
        with pdfplumber.open(filepath) as pdf:
            for page in pdf.pages:
                text += page.extract_text() + "\n"
    elif filepath.endswith(".docx"):

        doc = docx.Document(filepath)
        text = "\n".join([para.text for para in doc.paragraphs])
    elif filepath.endswith(".txt"):
        with open(filepath, "r", encoding="utf-8") as f:
            text = f.read()
    return text
```

#âœŒï¸ Function to Process Resumes

```python
import os

def process_resumes(drive_base_folder, cleaned_folder, overwrite_cleaned=False):
    """
    Processes all resumes in a folder (including subfolders), extracts text, cleans it, and saves the cleaned versions.

    Parameters:
    - drive_base_folder (str): The base folder where resumes are stored.
    - cleaned_folder (str): The destination folder to save cleaned resumes.
    - overwrite_cleaned (bool): If False, it skips files that are already cleaned.
    """

    print("âœ… Resume processing started!")
    print(f"Checking files in folder: {drive_base_folder}")
    print(f"Cleaned Resumes will be saved here: {cleaned_folder}")
    cleaned_count = 0

    for root, dirs, files in os.walk(drive_base_folder):
        for filename in files:
            # âœ… Skip non-resume files (only process .pdf, .docx, .txt)
            if not filename.endswith((".pdf", ".docx", ".txt")):
                continue

            # âœ… Extract text from the resume
            filepath = os.path.join(root, filename)
            try:
                raw_text = extract_text(filepath)
            except Exception as e:
                print(f"Error processing {filename}: {e}")
                continue

            # âœ… Clean the text (Define `clean_text()` elsewhere in your script)
            cleaned_text = clean_text(raw_text, remove_stopwords=False)

            # âœ… Define cleaned filename and save location
            if "_cleaned" not in filename:
                cleaned_filename = os.path.splitext(filename)[0] + "_cleaned.txt"
            else:
                cleaned_filename = filename  # Keep the original cleaned filename

            cleaned_filepath = os.path.join(cleaned_folder, cleaned_filename)

            # âœ… Handle overwriting logic
            if os.path.exists(cleaned_filepath) and not overwrite_cleaned:
                print(f"Skipped (already cleaned): {cleaned_filename}")
                continue

            # âœ… Write cleaned text to Google Drive
            with open(cleaned_filepath, "w", encoding="utf-8") as f:
                f.write(cleaned_text)
            print(f"Processed and saved: {cleaned_filename}")
            cleaned_count += 1

    print("âœ… All resumes have been cleaned and saved!")
    print(f"âœ… Total resumes cleaned: {cleaned_count}")
```

```python
import os
import pandas as pd
```

```python
# Clean text (remove special characters, lowercase, stopwords)
def additional_processing(text):
    """Cleans text by removing stopwords, special characters, and normalizing case."""
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"[^a-z0-9\s]", " ", text)  # Remove special characters
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
    doc = nlp(text)
    cleaned_text = " ".join([token.text for token in doc if not token.is_stop])  # Remove stopwords
    return cleaned_text
```

```python
  # Process all files in a given folder
def process_folder(folder_path, label):
    """Processes all text files in a folder and assigns a label."""
    resume_data = []

    if not os.path.exists(folder_path):
        print(f"âš ï¸ Warning: Folder not found: {folder_path}")
        return resume_data  # Return empty list if folder does not exist

    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):  # Only process text files
            file_path = os.path.join(folder_path, filename)

            with open(file_path, "r", encoding="utf-8") as file:
                text = file.read()

            cleaned_text = clean_text(text)  # Apply text cleaning
            name = os.path.splitext(filename)[0]  # Extract filename without extension
            resume_data.append([name, cleaned_text, label])  # Store data

    return resume_data
```

```python
# Save data to CSV
def save_to_csv(data, output_path):
    """Saves processed resume data to a CSV file."""
    df = pd.DataFrame(data, columns=["Name", "Text", "Label"])
    df.to_csv(output_path, index=False, encoding="utf-8")
    print(f"âœ… Successfully saved {len(df)} resumes to: {output_path}")
```

```python
# Main Execution
def merge_cleaned_resumes():
    """Processes and merges cleaned resumes from multiple folders into a single CSV."""

    print("ğŸš€ Starting the resume processing pipeline...")

    # Mount Google Drive
    # print("ğŸ“‚ Mounting Google Drive...")
    # mount_drive()
    print("âœ… Google Drive mounted successfully!\n")

    # Define source directories and labels
    base_dir = "/content/drive/My Drive/CAPSTONE/cleaned_colab_data"
    folders = {
        "cleaned_resumes_0": 0,  # Authentic resumes
        "cleaned_resumes_1": 1   # Non-authentic resumes
    }

    output_csv = "/content/drive/My Drive/CAPSTONE/cleaned_colab_data/cleaned_merged_resumes.csv"

    all_resumes = []

    # Process each folder
    for folder_name, label in folders.items():
        folder_path = os.path.join(base_dir, folder_name)

        print(f"ğŸ” Processing folder: {folder_name} (Label: {label})...")
        processed_resumes = process_folder(folder_path, label)

        if processed_resumes:
            print(f"âœ… {len(processed_resumes)} resumes processed in {folder_name}\n")
        else:
            print(f"âš ï¸ No resumes found in {folder_name}\n")

        all_resumes.extend(processed_resumes)

    # Save final dataset to CSV
    print(f"ğŸ’¾ Saving processed resumes to CSV at: {output_csv}...")
    save_to_csv(all_resumes, output_csv)

    print("ğŸ‰ Pipeline complete! Merged resumes saved successfully.\n")
```

```python
import pandas as pd
import re
import spacy

# Load SpaCy English model
nlp = spacy.load("en_core_web_sm")

def clean_resume_text(csv_path, output_csv_path):
    """Cleans the 'text' column in a CSV of resumes and saves the updated CSV."""

    print(f"ğŸ“‚ Loading CSV from: {csv_path}")
    df = pd.read_csv(csv_path)

    if "Text" not in df.columns:
        raise ValueError("The CSV does not contain a 'text' column.")

    print("ğŸ” Cleaning each resume in the 'text' column...")

    def clean_text(text):
        # Normalize case
        text = text.lower()

        # Remove special characters and extra whitespace
        text = re.sub(r"[^a-z0-9\s]", " ", text)
        text = re.sub(r"\s+", " ", text).strip()

        # Tokenize and remove stopwords using SpaCy
        doc = nlp(text)
        cleaned_tokens = [token.text for token in doc if not token.is_stop]

        return " ".join(cleaned_tokens)

    # Apply cleaning function to each resume
    df["Text"] = df["Text"].astype(str).apply(clean_text)

    # Save to CSV
    print(f"ğŸ’¾ Saving cleaned resumes to: {output_csv_path}")
    df.to_csv(output_csv_path, index=False)

    print("ğŸ‰ Done! Cleaned CSV saved.")
    return df
```

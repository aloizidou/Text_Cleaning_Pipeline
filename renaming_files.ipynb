# CODE TO RE-NAME FILES

### Run the following code to make chatGPT file names match the rest. 

```python
import os

#chatGPT
chatgpt_folder = "/Users/andrealoizidou/Desktop/capstone project/data_generation/chatGPT/job5"

print("Script started!")
print(f"Checking files in folder: {chatgpt_folder}")

for filename in os.listdir(chatgpt_folder):
    print(f"Checking file: {filename}")
    # Check if it matches the criteria
    starts_with_job = filename.startswith("job")
    contains_chatgpt = "chatgpt" in filename.lower()
    
    # Debug the matching logic
    print(f"  Starts with 'job': {starts_with_job}")
    print(f"  Contains 'chatGPT': {contains_chatgpt}")
    
    if starts_with_job and contains_chatgpt:
        parts = filename.split("_")
        job = parts[0]  # e.g., "job1"
        resume_number = parts[2]  # e.g., "7"
        job_number = job.replace("job", "")
        new_filename = f"{resume_number}_resume_job{job_number}_ChatGPT.pdf"

        src = os.path.join(chatgpt_folder, filename)
        dst = os.path.join(chatgpt_folder, new_filename)
        os.rename(src, dst)
        print(f"Renamed: {filename} -> {new_filename}")
    else:
        print(f"Skipped: {filename} (does not match renaming criteria)")

print("Renaming complete!")
```

## Run the following code if you deleted some files and you need to  RE-NUMBER the files.
> TWO FOR LOOPS (so that we can see the change)


```python
import os
import re

# Define the folder path
resume_folder = "/Users/andrealoizidou/Desktop/capstone project/authentic_resumes/job5"

print("Script started!")
print(f"Checking files in folder: {resume_folder}")

# Get and sort files, ignoring hidden system files like .DS_Store
file_list = sorted(
    [f for f in os.listdir(resume_folder) if os.path.isfile(os.path.join(resume_folder, f)) and not f.startswith('.')]
)

### STEP 1: Add an index at the beginning of each filename ###
for index, filename in enumerate(file_list, start=1):
    file_path = os.path.join(resume_folder, filename)
    
    # Construct new filename with index at the beginning
    new_filename = f"{index}_{filename}"
    new_file_path = os.path.join(resume_folder, new_filename)

    # Rename the file
    os.rename(file_path, new_file_path)
    print(f"Step 1 Renamed: {filename} -> {new_filename}")

# Get updated file list after step 1
updated_file_list = sorted(os.listdir(resume_folder))

### STEP 2: Remove everything between the first and second underscore ###
for filename in updated_file_list:
    file_path = os.path.join(resume_folder, filename)

    # Match the pattern: keep the first index and everything after the second underscore
    new_filename = re.sub(r'(_[^_]+_)', '_', filename, count=1)  # Removes text between first and second underscore

    new_file_path = os.path.join(resume_folder, new_filename)

    # Rename the file
    os.rename(file_path, new_file_path)
    print(f"Step 2 Renamed: {filename} -> {new_filename}")

print("Renaming complete!")
```

# Run following code if you quickly need to remove the last element. 
> e.g. 81_resume_job5_authentic_0.pdf_0 -> 81_resume_job5_authentic_0.pdf

```python
resume_folder = "/Users/andrealoizidou/Desktop/capstone project/authentic_resumes/job5"

print("Script started!")
print(f"Checking files in folder: {resume_folder}")

# Loop through all files in the folder
for filename in os.listdir(resume_folder):
    file_path = os.path.join(resume_folder, filename)

    # Skip non-files (like .DS_Store)
    if not os.path.isfile(file_path) or filename == ".DS_Store":
        continue

    # Remove the LAST occurrence of "_0" in the filename
    if filename.endswith("_0"):
        new_filename = filename.rsplit("_0", 1)[0]  # Remove last "_0"

        new_file_path = os.path.join(resume_folder, new_filename)

        # Rename the file
        os.rename(file_path, new_file_path)
        print(f"Renamed: {filename} -> {new_filename}")

print("Renaming complete!")
```

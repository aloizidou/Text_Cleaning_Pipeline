```python
import os
import shutil

# Define source and destination directories
source_dir = "/Users/andrealoizidou/Desktop/capstone_project/cleaned_resumes/cleaned_resumes_0"
destination_dir = "/Users/andrealoizidou/Desktop/capstone_project/UPDATED_cleaned_resume/cleaned_resumes_0"

# Ensure the destination directory exists
os.makedirs(destination_dir, exist_ok=True)

# Define prefixes to search for
prefixes = [
    "resume_job5_authentic_0_cleaned",
    "resume_job4_authentic_0_cleaned",
    "resume_job3_authentic_0_cleaned",
    "resume_job2_authentic_0_cleaned",
    "resume_job1_authentic_0_cleaned"
]

# Dictionary to store selected files for each prefix
selected_files = {prefix: [] for prefix in prefixes}

# Get all files in the source directory
all_files = sorted(os.listdir(source_dir))  # Sort for consistency

# Step 1: Select the first 150 for each category
for filename in all_files:
    for prefix in prefixes:
        if prefix in filename and len(selected_files[prefix]) < 150:
            selected_files[prefix].append(filename)
            break  # Stop checking after adding to one category

# Step 2: Select **11 additional unique files** for `resume_job5_authentic_0_cleaned` and `resume_job4_authentic_0_cleaned`
extra_files_needed = {"resume_job5_authentic_0_cleaned": 11, "resume_job4_authentic_0_cleaned": 11}

for filename in all_files:
    for prefix in extra_files_needed.keys():
        if (
            prefix in filename
            and filename not in selected_files[prefix]  # Ensure it's not already selected
            and len(selected_files[prefix]) < 161  # Total should be 150 + 11
        ):
            selected_files[prefix].append(filename)

# Verify selection
total_selected = sum(len(files) for files in selected_files.values())

# Copy selected files to the new directory
for prefix, files in selected_files.items():
    for file in files:
        source_path = os.path.join(source_dir, file)
        dest_path = os.path.join(destination_dir, file)
        shutil.copy2(source_path, dest_path)  # Copy with metadata

# Print verification statement
print(f"✅ Successfully selected and copied {total_selected} files to {destination_dir}.")

# Print counts per category
for prefix, files in selected_files.items():
    print(f"{prefix}: {len(files)} files selected.")

# on Purpose 772!!!
# Ensure the total is exactly 772 (750 + 22 additional)
if total_selected == 772:
    print("✅ Total files correctly selected: 772")
else:
    print(f"⚠️ Warning: Selected {total_selected} files instead of 772!")
```

# More Preprocessing
•	Remove special characters, extra whitespace, and non-alphanumeric symbols.

•	Normalize case by converting all text to lowercase.

•	Optionally remove stopwords (common words like "the," "is," etc.) unless they are contextually significant in resumes.

```python
import os
import re
import nltk
from nltk.corpus import stopwords

# Ensure stopwords are downloaded
nltk.download("stopwords")

# Define source directories
base_dir = "/Users/andrealoizidou/Desktop/capstone_project/UPDATED_cleaned_resume"
source_dirs = [os.path.join(base_dir, "cleaned_resumes_1"), os.path.join(base_dir, "cleaned_resumes_0")]

# Define destination directory
destination_dir = os.path.join(base_dir, "removed_stopwords")
os.makedirs(destination_dir, exist_ok=True)

# Load stopwords
stop_words = set(stopwords.words("english"))

def clean_resume(text, remove_stopwords=True):
    """
    Cleans the resume text by:
    - Removing special characters and extra whitespace
    - Converting text to lowercase
    - Optionally removing stopwords
    """
    # Remove special characters and non-alphanumeric symbols (keep spaces)
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)

    # Convert to lowercase
    text = text.lower()

    # Remove extra spaces
    text = " ".join(text.split())

    # Remove stopwords if required
    if remove_stopwords:
        text = " ".join([word for word in text.split() if word not in stop_words])

    return text

# Process resumes
for source_dir in source_dirs:
    if not os.path.exists(source_dir):
        print(f"⚠️ Warning: Directory not found - {source_dir}")
        continue

    for filename in os.listdir(source_dir):
        source_path = os.path.join(source_dir, filename)

        # Skip non-text files
        if not filename.endswith(".txt"):
            continue

        # Read the resume text
        with open(source_path, "r", encoding="utf-8") as file:
            resume_text = file.read()

        # Clean the resume
        cleaned_text = clean_resume(resume_text)

        # Save the cleaned resume
        dest_path = os.path.join(destination_dir, filename)
        with open(dest_path, "w", encoding="utf-8") as file:
            file.write(cleaned_text)

        print(f"✅ Processed and saved: {filename}")

print(f" All cleaned resumes are saved in: {destination_dir}")
```

# Generating CSV file
> Assigns labels – Resumes containing "authentic" in their filename are labeled as 0 (authentic), while others are labeled as 1 (non-authentic).

> Columns: Name, Text, and Label.

- Saves the DataFrame as a CSV file in the specified path.

```python
import os
import pandas as pd

# Define source and destination directories
source_dir = "/Users/andrealoizidou/Desktop/capstone_project/UPDATED_cleaned_resume/removed_stopwords"
csv_path = "/Users/andrealoizidou/Desktop/capstone_project/UPDATED_cleaned_resume/data_generation.csv"

# Ensure the source directory exists before proceeding
if not os.path.exists(source_dir):
    print(f"❌ Error: Source directory not found: {source_dir}")
    exit(1)

# List to store resume data
resume_data = []

# Process each file in the directory
for filename in os.listdir(source_dir):
    if filename.endswith(".txt"):  # Ensure it's a text file
        file_path = os.path.join(source_dir, filename)

        # Read the file content
        with open(file_path, "r", encoding="utf-8") as file:
            text = file.read()

        # Extract the name (without .txt extension)
        name = os.path.splitext(filename)[0]

        # Assign label (0 for authentic, 1 for non-authentic)
        label = 0 if "authentic" in filename.lower() else 1

        # Append to the list
        resume_data.append([name, text, label])

# Convert to a DataFrame
df = pd.DataFrame(resume_data, columns=["Name", "Text", "Label"])

# Save as CSV
df.to_csv(csv_path, index=False, encoding="utf-8")

# Print verification message
print(f"✅ Successfully created CSV file: {csv_path}")
print(f"📄 Total resumes processed: {len(df)}")
```
